```python
df_seeds = pd.read_csv('../resources/data/unprocessed/MNCAATourneySeeds.csv')

df_seeds = df_seeds.loc[df_seeds['Season'] == 2023, :].drop(columns=['Season']).reset_index(drop=True)

df_seeds['PlayIn'] = 0
df_seeds.loc[df_seeds['Seed'].str[-1].str.isalpha(), 'PlayIn'] = 1

df_seeds['Region'] = df_seeds['Seed'].str[0]

df_seeds['Seed'] = df_seeds['Seed'].str.extract('(\d+)').astype(int)

df_seeds
```

```python
def create_bracket_list(df_seeds=df_seeds):
    round0 = []
    region = ['W', 'X', 'Y', 'Z']
    for r in region:
        # print(r)
        for s in [1, 8, 5, 4, 6, 3, 7, 2]:
            a = df_seeds.loc[(df_seeds['Region'] == r) & (df_seeds['Seed'] == s), 'TeamID']
            if a.shape[0] == 1: round0.append([a.iloc[0]])
            else: round0.append([a.iloc[0], a.iloc[1]])
            
            b = df_seeds.loc[(df_seeds['Region'] == r) & (df_seeds['Seed'] == 17 - s), 'TeamID']
            if b.shape[0] == 1: round0.append([b.iloc[0]])
            else: round0.append([b.iloc[0], b.iloc[1]])
            
            # print()
    return round0
            
round0 = create_bracket_list()

# remove playins
round0[1] = [1192]
round0[17] = [1394]
round0[41] = [1338]
round0[-7] = [1113]

# reorder to match ESPN
round0 = round0[16:32] + round0[:16] + round0[32:64]

teams = [i[0] for i in round0]

teams
```

```python
def get_round_survivors(teams, probs, used_teams):        
    new_teams = set()
    for index, pool in enumerate(teams):
        intersect = used_teams.intersection(pool)
        if len(intersect) == 0:
            pool_probs = np.array(probs[index, :])
            new_teams.add(np.random.choice(pool, p=pool_probs))
    
    return new_teams

get_round_survivors(round_teams[6], round_model_probs[6], set())
```

```python
def create_bracket(round_probs, round_teams=round_teams): 
    # print('CHAMPION')
    used_teams = get_round_survivors(round_teams[6], round_probs[6], set())
    bracket = dict(zip(used_teams, np.repeat(6, 1)))
    # print(survivors)
    # print()
    
    # print('CHAMPIONSHIP')    
    new_teams = get_round_survivors(round_teams[5], round_probs[5], used_teams)
    bracket.update(dict(zip(new_teams, np.repeat(5, 1))))
    used_teams.update(new_teams)
    # print(survivors)
    # print()
    
    # print('FINAL FOUR')    
    new_teams = get_round_survivors(round_teams[4], round_probs[4], used_teams)
    bracket.update(dict(zip(new_teams, np.repeat(4, 2))))
    used_teams.update(new_teams)
    # print(survivors)
    # print()
    
    # print('ELITE EIGHT')    
    new_teams = get_round_survivors(round_teams[3], round_probs[3], used_teams)
    bracket.update(dict(zip(new_teams, np.repeat(3, 4))))
    used_teams.update(new_teams)
    # print(survivors)
    # print()
    
    # print('SWEET SIXTEEN')    
    new_teams = get_round_survivors(round_teams[2], round_probs[2], used_teams)
    bracket.update(dict(zip(new_teams, np.repeat(2, 8))))
    used_teams.update(new_teams)
    # print(survivors)
    # print()
    
    # print('ROUND OF 32')    
    new_teams = get_round_survivors(round_teams[1], round_probs[1], used_teams)
    bracket.update(dict(zip(new_teams, np.repeat(1, 16))))
    used_teams.update(new_teams)
    # print(survivors)
    # print()
    
    return bracket

create_bracket(round_model_probs)
```

```python
def score_bracket(actual, predicted):
    score = 0
    for team, predicted_finish in predicted.items():
        if actual_placement := actual.get(team):
            if actual_placement < predicted_finish:
                predicted_finish = actual_placement
                
            if predicted_finish == 1:
                score += 10
            elif predicted_finish == 2:
                score += 30
            elif predicted_finish == 3:
                score += 70
            elif predicted_finish == 4:
                score += 150
            elif predicted_finish == 5:
                score += 310
            elif predicted_finish == 6:
                score += 630
                
    return score

score_bracket(public_bracket, actual_results)
```

```python
def bracket_combo(index1, bracket1, potential_brackets_payout_top_1000=potential_brackets_payout_top_1000):
    index2 = np.argmax(potential_brackets_payout_top_1000[bracket1 == 0, :].mean(axis=0))
    expected_payout = bracket1.mean() + potential_brackets_payout_top_1000[:, index2].mean()
    
    return (index1, index2, expected_payout)

bracket_combo(84, potential_brackets_payout_top_1000[:, 84])
```
